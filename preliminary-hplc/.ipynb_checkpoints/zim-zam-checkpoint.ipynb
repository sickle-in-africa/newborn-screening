```python
#!/usr/bin/env python3
import pandas as pd

# # Load data
# data_zim = pd.read_csv('data/zim_zam/phase_0/raw/zim.csv')
# data_zam = pd.read_csv('data/zim_zam/phase_0/raw/zam.csv')

# Standardize column names by stripping spaces and replacing them with underscores
data_zam.columns = data_zam.columns.str.strip().str.replace(' ', '_')
data_zim.columns = data_zim.columns.str.strip().str.replace(' ', '_')

# Clean zam data
data_zam_clean = (
    data_zam
    .assign(
        Facility_ID='',
        country='zim/zam',
        StudyID=data_zam['StudyID'].astype(str)  # Ensure StudyID is string
    )
    .replace('', pd.NA)  # Replace empty strings with NA
    .loc[:, ['StudyID', 'Facility_ID', 'country', 'Sex', 'Standard_POCT_Results', 'DBS_POCT_Results', 'IEF_Results']]
)

# Clean zim data
data_zim_clean = (
    data_zim
    .assign(
        country='zim/zam',
        StudyID=data_zim['StudyID'].astype(str)  # Ensure StudyID is string
    )
    .replace('', pd.NA)  # Replace empty strings with NA
    .loc[:, ['StudyID', 'Facility_ID', 'country', 'Sex', 'Standard_POCT_Results', 'DBS_POCT_Results', 'IEF_Results']]
)

# Combine the two datasets
data_zim_zam_clean = pd.concat([data_zam_clean, data_zim_clean], ignore_index=True)



data_zim_zam_clean.head()
```

```python
# Value counts and percentages for the 'Sex' column
sex_counts = data_zim_zam_clean['Sex'].value_counts(dropna=False)
sex_percentage = (data_zim_zam_clean['Sex'].value_counts(dropna=False, normalize=True) * 100).round(2)

# Combine counts and percentages into a single DataFrame
sex_summary = pd.DataFrame({
    'Count': sex_counts,
    'Percentage': sex_percentage
})

# Print the result
print(sex_summary)
```

```python
# Define result mappings
AA_ief = ["FA - NO abnormal hemoglobin"]
AA_ief1 = ["AF - NO abnormal hemoglobin"]
AA_ief2 = ["Fa"]
AA_ief3 = ["FA"]
SS_ief = ["FS - Sickle cell disease SS (SCD-SS)/Sickle cell disease beta-0-thalassemia"]
SS_ief1 = ["FS"]
AS_ief = ["FAS - Sickle cell trait (AS)"]
AS_ief1 = ["FAS"]
```

```python
# Standardize 'result_gs' values
result_gs_mapping = {
    **{k: 'AA' for k in AA_ief + AA_ief1 + AA_ief2 + AA_ief3},
    **{k: 'SS' for k in SS_ief + SS_ief1},
    **{k: 'AS' for k in AS_ief + AS_ief1}
}

# Replace 'result_gs' values based on mapping
data_zim_zam_clean['IEF_Results'] = data_zim_zam_clean['IEF_Results'].replace(result_gs_mapping)
```
