```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Actual enrollment data (Zambia + Zimbabwe combined)
actual_data = {
    "Country": ["Ghana", "Mali", "Nigeria", "Tanzania", "Uganda", "Zimbabwe/Zambia"],
    "Actual": [999, 1115, 2407, 900, 980, 550 + 450]  # combine Zimbabwe + Zambia
}

# Target enrollment
target_data = {
    "Ghana": 1000,
    "Mali": 1000,
    "Nigeria": 2500,
    "Tanzania": 1000,
    "Uganda": 1000,
    "Zimbabwe/Zambia": 1000
}

df = pd.DataFrame(actual_data)
df["Target"] = df["Country"].map(target_data)

# ---- Plot ----
x = np.arange(len(df["Country"]))  # positions
width = 0.35  # bar width

fig, ax = plt.subplots(figsize=(12, 6))

bars1 = ax.bar(x - width/2, df["Actual"], width, label="Actual")
bars2 = ax.bar(x + width/2, df["Target"], width, label="Target")

# Annotate counts above bars
for rect, cnt in zip(bars1, df["Actual"]):
    ax.text(rect.get_x() + rect.get_width()/2, rect.get_height() + 20,
            f"{cnt}", ha="center", va="bottom", fontsize=10)

for rect, cnt in zip(bars2, df["Target"]):
    ax.text(rect.get_x() + rect.get_width()/2, rect.get_height() + 20,
            f"{cnt}", ha="center", va="bottom", fontsize=10, color="black")

# Clean axes
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_ylabel("Enrollment Counts", fontsize=12)
ax.set_title("Newborn Screening Enrollment: Actual vs Target")

# Increase x-axis font size
ax.set_xticks(x)
ax.set_xticklabels(df["Country"], fontsize=12)
# ax.set_yticklabels(df["Country"], fontsize=12)

# Legend without border
legend = ax.legend(frameon=False)

plt.tight_layout()
plt.savefig("enrollment_actual_vs_target_clean.png", dpi=600, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------- Helper function ----------
def make_country_df(rows):
    items, sd, d, n, a, sa = [], [], [], [], [], []
    for item, vals in rows.items():
        sdis, dis, neu, agr, sagr, miss = vals
        total_nonmiss = sdis + dis + neu + agr + sagr
        items.append(item)
        sd.append(100 * sdis / total_nonmiss)
        d.append(100 * dis  / total_nonmiss)
        n.append(100 * neu  / total_nonmiss)
        a.append(100 * agr  / total_nonmiss)
        sa.append(100 * sagr / total_nonmiss)
    return pd.DataFrame({
        "Item": items,
        "Strongly Disagree": sd,
        "Disagree": d,
        "Neutral": n,
        "Agree": a,
        "Strongly Agree": sa
    })

# ---------- (example) combine Zambia + Zimbabwe ----------
# (reuse your data dicts from before)
zz = {k: [a+b for a, b in zip(zambia[k], zimbabwe[k])] for k in zambia.keys()}

# Build tidy DataFrames
df_mali     = make_country_df(mali)
df_nigeria  = make_country_df(nigeria)
df_tanzania = make_country_df(tanzania)
df_uganda   = make_country_df(uganda)
df_zz       = make_country_df(zz)

# Common order with scd_detect_birth at the TOP
order = [
    "scd_detect_birth", "scd_live_long", "scd_early_detection",
    "test_result_immediate", "poc_info_access", "poc_faster_result",
    "recommend_scd_screening", "recommend_poc", "dbs_acceptable", "seek_treatment"
]
def reorder(df):
    out = df.set_index("Item").loc[order]
    return out.iloc[::-1]

dfs = {
    "Mali": reorder(df_mali),
    "Nigeria": reorder(df_nigeria),
    "Tanzania": reorder(df_tanzania),
    "Uganda": reorder(df_uganda),
    "Zimbabwe/Zambia": reorder(df_zz)
}

# ---------- Plot with custom layout ----------
fig, axes = plt.subplots(2, 3, figsize=(17, 10))
axes = axes.ravel()

likert_cols = ["Strongly Disagree", "Disagree", "Neutral", "Agree", "Strongly Agree"]
colors = ["#d73027", "#fc8d59", "#ffffbf", "#91bfdb", "#4575b4"]

for i, (ax, (country, dff)) in enumerate(zip(axes, dfs.items())):
    dff[likert_cols].plot(kind="barh", stacked=True, ax=ax, color=colors, legend=False)
    ax.set_title(country, fontsize=13)
    ax.set_xlabel("Percentage")
    
    # Show y-labels only on first plot of each row
    if i not in [0, 3]:  
        ax.set_yticklabels([])
        ax.set_ylabel("")
    else:
        ax.set_ylabel("Survey Item")

    # Clean borders
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

# Hide the unused 6th subplot
axes[-1].axis("off")

# Center bottom row (Uganda + Zimbabwe/Zambia)
axes[3].set_position([0.20, 0.08, 0.25, 0.38])   # left bottom plot
axes[4].set_position([0.55, 0.08, 0.25, 0.38])   # right bottom plot

# Global legend
fig.legend(likert_cols, loc="lower center", ncol=5, frameon=False, fontsize=11)

fig.suptitle("Survey Responses by Country (Likert %)\n(Zimbabwe and Zambia combined)", y=0.96, fontsize=16)
plt.tight_layout(rect=[0,0.08,1,0.95])
plt.savefig("likert_subplots_countries_centered.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ---- helper to create % DataFrame ----
def make_country_df(rows):
    items, sd, d, n, a, sa = [], [], [], [], [], []
    for item, vals in rows.items():
        sdis, dis, neu, agr, sagr, miss = vals
        total_nonmiss = sdis + dis + neu + agr + sagr
        items.append(item)
        sd.append(100 * sdis / total_nonmiss)
        d.append(100 * dis  / total_nonmiss)
        n.append(100 * neu  / total_nonmiss)
        a.append(100 * agr  / total_nonmiss)
        sa.append(100 * sagr / total_nonmiss)
    return pd.DataFrame({
        "Item": items,
        "Strongly Disagree": sd,
        "Disagree": d,
        "Neutral": n,
        "Agree": a,
        "Strongly Agree": sa
    })

# ---- combine Zambia + Zimbabwe ----
zz = {k: [a+b for a, b in zip(zambia[k], zimbabwe[k])] for k in zambia.keys()}

df_mali     = make_country_df(mali)
df_nigeria  = make_country_df(nigeria)
df_tanzania = make_country_df(tanzania)
df_uganda   = make_country_df(uganda)
df_zz       = make_country_df(zz)

# ---- Helper reorder function ----
order = [
    "scd_detect_birth", "scd_live_long", "scd_early_detection",
    "test_result_immediate", "poc_info_access", "poc_faster_result",
    "recommend_scd_screening", "recommend_poc", "dbs_acceptable", "seek_treatment"
]

def reorder(df):
    out = df.set_index("Item").loc[order]
    return out.iloc[::-1]

dfs = {
    "Mali": reorder(df_mali),
    "Nigeria": reorder(df_nigeria),
    "Tanzania": reorder(df_tanzania),
    "Uganda": reorder(df_uganda),
    "Zimbabwe/Zambia": reorder(df_zz)
}

likert_cols = ["Strongly Disagree", "Disagree", "Neutral", "Agree", "Strongly Agree"]
colors = ["#d73027", "#fc8d59", "#ffffbf", "#91bfdb", "#4575b4"]

# ---- Custom GridSpec: 3 cols on top, 2 centered on bottom ----
fig = plt.figure(figsize=(17, 10))
gs = gridspec.GridSpec(2, 3, height_ratios=[1, 1])

# Top row (3 plots)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[0,2])

# Bottom row (Uganda between col0 & col1, Zimbabwe/Zambia between col1 & col2)
ax4 = fig.add_subplot(gs[1,0:2])  # Uganda
ax5 = fig.add_subplot(gs[1,1:3])  # Zimbabwe/Zambia

axes = [ax1, ax2, ax3, ax4, ax5]
countries = list(dfs.keys())

# ---- Plot each country ----
for i, (ax, country) in enumerate(zip(axes, countries)):
    dff = dfs[country]
    dff[likert_cols].plot(kind="barh", stacked=True, ax=ax, color=colors, legend=False)
    ax.set_title(country, fontsize=13)
    ax.set_xlabel("Percentage")
    
    # Show y-axis labels only on far-left plots (ax1 and ax4)
    if ax not in [ax1, ax4]:
        ax.set_yticklabels([])
        ax.set_ylabel("")
    else:
        ax.set_ylabel("Survey Item")
    
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

# Global legend
fig.legend(likert_cols, loc="lower center", ncol=5, frameon=False, fontsize=11)

fig.suptitle("Survey Responses by Country (Likert %)\n(Zimbabwe and Zambia combined)", y=0.96, fontsize=16)
plt.tight_layout(rect=[0,0.08,1,0.95])
plt.savefig("likert_subplots_centered.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# ---------- Helper: convert raw counts to % for Likert categories (exclude Missing) ----------
def make_country_df(rows):
    items, sd, d, n, a, sa = [], [], [], [], [], []
    for item, vals in rows.items():
        sdis, dis, neu, agr, sagr, miss = vals
        total = sdis + dis + neu + agr + sagr  # exclude missing
        items.append(item)
        sd.append(100 * sdis / total if total else 0)
        d.append(100 * dis  / total if total else 0)
        n.append(100 * neu  / total if total else 0)
        a.append(100 * agr  / total if total else 0)
        sa.append(100 * sagr / total if total else 0)
    return pd.DataFrame({
        "Item": items,
        "Strongly Disagree": sd,
        "Disagree": d,
        "Neutral": n,
        "Agree": a,
        "Strongly Agree": sa
    })

# ---------- Raw data you provided ----------
mali = {
    "dbs_acceptable":          [4,10,273,814,11,3],
    "poc_faster_result":       [3,12,222,861,16,1],
    "poc_info_access":         [1,15,167,866,66,0],
    "recommend_poc":           [2,9,181,873,49,1],
    "recommend_scd_screening": [2,62,129,881,41,0],
    "scd_detect_birth":        [10,80,278,712,23,12],
    "scd_early_detection":     [4,21,277,782,31,0],
    "scd_live_long":           [2,22,332,744,14,1],
    "seek_treatment":          [1,67,105,849,80,13],
    "test_result_immediate":   [2,21,236,845,11,0],
}

nigeria = {
    "dbs_acceptable":          [11,41,542,1602,201,10],
    "poc_faster_result":       [7,67,592,1534,201,6],
    "poc_info_access":         [14,135,443,1645,161,9],
    "recommend_poc":           [8,18,250,1752,365,14],
    "recommend_scd_screening": [12,18,261,1694,418,4],
    "scd_detect_birth":        [35,146,595,1359,269,3],
    "scd_early_detection":     [9,46,319,1716,307,10],
    "scd_live_long":           [47,320,443,1392,198,7],
    "seek_treatment":          [11,7,229,1741,409,10],
    "test_result_immediate":   [8,49,389,1721,226,14],
}

tanzania = {
    "dbs_acceptable":          [0,4,80,709,107,0],
    "poc_faster_result":       [0,9,183,589,119,0],
    "poc_info_access":         [0,5,40,726,129,0],
    "recommend_poc":           [0,1,52,754,93,0],
    "recommend_scd_screening": [0,3,41,706,150,0],
    "scd_detect_birth":        [3,38,252,543,64,0],
    "scd_early_detection":     [0,6,103,659,132,0],
    "scd_live_long":           [4,68,218,523,87,0],
    "seek_treatment":          [1,2,35,724,138,0],
    "test_result_immediate":   [0,6,88,675,131,0],
}

uganda = {
    "dbs_acceptable":          [0,9,168,603,193,7],
    "poc_faster_result":       [0,9,134,568,263,6],
    "poc_info_access":         [29,138,79,466,265,3],
    "recommend_poc":           [0,3,63,594,314,6],
    "recommend_scd_screening": [0,8,68,566,331,7],
    "scd_detect_birth":        [6,15,135,289,531,4],
    "scd_early_detection":     [0,13,100,496,364,7],
    "scd_live_long":           [14,93,81,570,219,3],
    "seek_treatment":          [1,6,65,633,265,10],
    "test_result_immediate":   [0,13,71,598,292,6],
}

zambia = {
    "dbs_acceptable":          [0,2,178,137,133,100],
    "poc_faster_result":       [0,2,127,180,141,100],
    "poc_info_access":         [3,21,178,120,128,100],
    "recommend_poc":           [0,1,31,218,200,100],
    "recommend_scd_screening": [0,0,14,189,247,100],
    "scd_detect_birth":        [0,7,145,134,164,100],
    "scd_early_detection":     [0,0,27,190,233,100],
    "scd_live_long":           [3,9,69,206,163,100],
    "seek_treatment":          [0,0,10,179,261,100],
    "test_result_immediate":   [0,0,53,196,201,100],
}

zimbabwe = {
    "dbs_acceptable":          [0,0,9,431,10,0],
    "poc_faster_result":       [0,0,14,424,12,0],
    "poc_info_access":         [0,1,19,420,10,0],
    "recommend_poc":           [0,0,2,423,25,0],
    "recommend_scd_screening": [0,0,1,408,41,0],
    "scd_detect_birth":        [0,0,5,405,40,0],
    "scd_early_detection":     [0,0,5,405,40,0],
    "scd_live_long":           [0,0,25,385,40,0],
    "seek_treatment":          [0,0,1,408,41,0],
    "test_result_immediate":   [0,0,2,400,48,0],
}

# ---------- Combine Zambia + Zimbabwe ----------
zz = {k: [a+b for a, b in zip(zambia[k], zimbabwe[k])] for k in zambia.keys()}

# ---------- Build % data frames ----------
df_mali     = make_country_df(mali)
df_nigeria  = make_country_df(nigeria)
df_tanzania = make_country_df(tanzania)
df_uganda   = make_country_df(uganda)
df_zz       = make_country_df(zz)

# Order items; reverse for barh so scd_detect_birth is at top
order = [
    "scd_detect_birth", "scd_live_long", "scd_early_detection",
    "test_result_immediate", "poc_info_access", "poc_faster_result",
    "recommend_scd_screening", "recommend_poc", "dbs_acceptable", "seek_treatment"
]
def reorder(df):
    out = df.set_index("Item").loc[order]
    return out.iloc[::-1]

dfs = {
    "Mali": reorder(df_mali),
    "Nigeria": reorder(df_nigeria),
    "Tanzania": reorder(df_tanzania),
    "Uganda": reorder(df_uganda),
    "Zimbabwe/Zambia": reorder(df_zz)
}

likert_cols = ["Strongly Disagree", "Disagree", "Neutral", "Agree", "Strongly Agree"]
colors = ["#d73027", "#fc8d59", "#ffffbf", "#91bfdb", "#4575b4"]

# ---------- Manual layout: top row (3 plots), bottom row (2 centered between gaps) ----------
fig = plt.figure(figsize=(17, 10))

# Top row axes (made wider: width = 0.3)
ax1 = fig.add_axes([0.05, 0.55, 0.3, 0.35])   # Mali
ax2 = fig.add_axes([0.38, 0.55, 0.3, 0.35])   # Nigeria
ax3 = fig.add_axes([0.71, 0.55, 0.3, 0.35])   # Tanzania

# Bottom row axes (wider: width = 0.3)
ax4 = fig.add_axes([0.22, 0.08, 0.3, 0.35])   # Uganda
ax5 = fig.add_axes([0.55, 0.08, 0.3, 0.35])   # Zimbabwe/Zambia

# # Top row axes (equal width)
# ax1 = fig.add_axes([0.05, 0.55, 0.25, 0.35])  # Mali
# ax2 = fig.add_axes([0.37, 0.55, 0.25, 0.35])  # Nigeria
# ax3 = fig.add_axes([0.69, 0.55, 0.25, 0.35])  # Tanzania

# # Bottom row axes: centered under gaps between top plots (narrower)
# ax4 = fig.add_axes([0.21, 0.08, 0.25, 0.35])  # Uganda (between Mali & Nigeria)
# ax5 = fig.add_axes([0.53, 0.08, 0.25, 0.35])  # Zimbabwe/Zambia (between Nigeria & Tanzania)

axes = [ax1, ax2, ax3, ax4, ax5]
countries = list(dfs.keys())

for ax, country in zip(axes, countries):
    dff = dfs[country]
    dff[likert_cols].plot(kind="barh", stacked=True, ax=ax, color=colors, legend=False)
    ax.set_title(country, fontsize=17)
    ax.set_xlabel("Percentage", fontsize=16, labelpad=10)
    ax.tick_params(axis="y", labelsize=15)
    # Show y-axis labels only on far-left of each row (Mali, Uganda)
    if country not in ["Mali", "Uganda"]:
        ax.set_yticklabels([], fontsize=17)
        ax.set_ylabel("")
    else:
        ax.set_ylabel("Survey Item", fontsize=16)
    # Clean borders
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

# One shared legend at the bottom
fig.legend(likert_cols, loc="lower center", ncol=5, frameon=False, fontsize=17, bbox_to_anchor=(0.5, -0.06))

fig.suptitle("Survey Responses by Country (Likert %)", y=0.98, fontsize=19)
plt.savefig("likert_subplots_centered_manual.png", dpi=300, bbox_inches="tight")
plt.show()
```
