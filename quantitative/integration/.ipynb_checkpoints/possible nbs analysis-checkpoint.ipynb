```python
import numpy as np
import pandas as pd
from statsmodels.miscmodels.ordinal_model import OrderedModel
from sklearn.preprocessing import OrdinalEncoder
```

#### Possible research question

_Does SCD detection at birth influence healthcare workers' opinion on early screening?_

#### Load and preprocess the data

Rename columns to shorter names

Deal with missing values (impute or remove them)

Convert survey questions to Likert scale

Identify ordinal columns

Encode ordinal values

```python
def convert_to_datetime(date_str):
    try:
        return pd.to_datetime(date_str, errors='coerce', dayfirst=True)
    except Exception as e:
        print(f"Error converting {date_str}: {e}")
        return NaT
```

```python
# define the ordinal categories using a Likert scale
likert_scale = ["Strongly Disagree", "Disagree", "Neutral", "Agree", "Strongly Agree"]

# identify ordinal columns 
ordinal_cols = [
    "Sickle cell disease (SCD) can be detected at birth",
    "People with SCD can live long lives",
    "Early detection of SCD is needed to provide adequate care for the affected child",
    "It is important that a test for SCD for my child produces the result immediately",
    "I have access to all the information I need on point of care tests for SCD",
    "The result of a point of care test for SCD is available faster than for other screening methods",
    "I would recommend the screening of babies for SCD at other healthcare facilities",
    "I would recommend the use of point of care tests for SCD at other healthcare facilities",
    "Alongside the point of care test, it is acceptable for a dried blood spot sample to be collected for validation",
    "If someone tests positive for SCD at this healthcare facility it is important that they seek proper medical treatment, either here or at another facility"
]

other_cols = ["record_id",
              "participant_id",
              "facility_id",
              "dob_guardian",
              "guardian_relationship",
              "no_followup_reasons",
              "country"]

# identify date columns
date_cols = ['dob_guardian']

mapping_sex = {"Male": 0, "Female": 1}
mapping_relation = {"Grandparent": 0,
                   "Father": 1,
                   "Mother": 2,
                   "Aunt": 3}
```

```python
# load the data
data = pd.read_excel("data/quantitative_clean_data.xlsx") 


# drop NaN's for now
data.dropna(inplace=True) 


# convert to datetime
data['dob_guardian'] = data['dob_guardian'].apply(convert_to_datetime)


# encode ordinal features
encoder = OrdinalEncoder(categories=[likert_scale] * len(ordinal_cols)) 
data[ordinal_cols] = encoder.fit_transform(data[ordinal_cols])
# encode the other features as well
data["sex_guardian"] = data["sex_guardian"].map(mapping_sex)
data["guardian_relationship"] = data["guardian_relationship"].map(mapping_relation)
data["country"] = data["country"].map({"Ghana":1})

# for hierarchical regression define several matrices of independent variables
X0 = data[["sex_guardian"]]  # control variable for building base model
X1 = data[["sex_guardian", 
      "guardian_relationship",  
      "Sickle cell disease (SCD) can be detected at birth"]] # model to answer research question above

# define dependent variable
y = data["I would recommend the screening of babies for SCD at other healthcare facilities"]

model0 = OrderedModel(y, X0, distr="logit")
result0 = model0.fit(method="bfgs")

model1 = OrderedModel(y, X1, distr="logit")
result1 = model1.fit(method="bfgs")

print("\nBaseline Model results: ")
print(result0.summary())

print("\nModel with additional features: ")
print(result1.summary())

```

### Interpretation

1. For baseline model, with p-value is $0.891$: `sex_guardian_Male` is NOT a significant predictor, so is not statistically significant.

2. `Sickle cell disease (SCD) can be detected at birth)` with p-value `0.000` which is highly signficant suggests people who believe SCD can be detected at birth are much more likely to recommend screening at other facilities.
