```python
import numpy as np
import scipy.stats as stats

def calculate_mean_ci(df, column, confidence=0.95):
    df_filtered = df[df[column] >= 0]
    
    mean_value_days = df_filtered[column].mean()
    n = len(df_filtered[column])
    std_err = stats.sem(df_filtered[column])
    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    
    ci_lower_days = mean_value_days - margin_of_error
    ci_upper_days = mean_value_days + margin_of_error
    
    mean_value_months = mean_value_days / 30.44  # Convert days to months (average month length)
    ci_lower_months = ci_lower_days / 30.44
    ci_upper_months = ci_upper_days / 30.44
    
    return mean_value_months, (ci_lower_months, ci_upper_months)
```

```python
def map_and_plot_columns_by_country(df, columns, value_map, figsize=(12, 5)):
    """
    Map columns to descriptive labels, create histograms for each country (plots side by side for each country),
    and include an overall summary plot at the end. Reverts back to numbers after plotting.

    Parameters:
    - df (DataFrame): The DataFrame to process.
    - columns (list): List of column names to map and plot.
    - value_map (dict): Mapping of numerical values to descriptive labels.
    - figsize (tuple): Figure size for each country's plots.

    Returns:
    - DataFrame: The original DataFrame with columns reverted back to numbers.
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    import matplotlib.ticker as ticker

    # Extract the unique list of countries
    countries = df['country'].unique()

    # Step 1: Map values to descriptive labels
    df_temp = df[columns + ['country']].astype('object')  # Include 'country' for mapping
    df_mapped = df_temp.replace(value_map)

    # Step 2: Generate plots for each country
    for country in countries:
        # Filter data for the current country
        country_data = df_mapped[df_mapped['country'] == country]

        # Create subplots for the current country
        fig, axes = plt.subplots(1, len(columns), figsize=figsize, sharey=True)

        for i, col in enumerate(columns):
            sns.histplot(country_data[col], kde=False, discrete=True, ax=axes[i], color=sns.color_palette("tab10")[i])
            axes[i].set_title(f"{col.replace('_', ' ').title()} in {country.title()}", fontsize=10)
            axes[i].set_xlabel('Response', fontsize=10)
            if i == 0:  # Add ylabel only for the first plot
                axes[i].set_ylabel('Frequency', fontsize=10)
            axes[i].tick_params(axis='x', rotation=45, labelsize=10)  # Rotate x-axis labels
            axes[i].tick_params(axis='y', labelsize=10)  # Adjust y-axis font size
            axes[i].spines['top'].set_visible(False)  # Remove top border
            axes[i].spines['right'].set_visible(False)  # Remove right border
            axes[i].grid(visible=False)  # Remove grid lines
            axes[i].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x)}' if x == int(x) else ''))

        plt.tight_layout()
        plt.show()

    # Step 3: Generate overall plots
    fig, axes = plt.subplots(1, len(columns), figsize=figsize, sharey=True)

    for i, col in enumerate(columns):
        sns.histplot(df_mapped[col], kde=False, discrete=True, ax=axes[i], color=sns.color_palette("pastel")[i])
        axes[i].set_title(f"Overall {col.replace('_', ' ').title()}", fontsize=10)
        axes[i].set_xlabel('Response', fontsize=10)
        if i == 0:  # Add ylabel only for the first plot
            axes[i].set_ylabel('Frequency', fontsize=10)
        axes[i].tick_params(axis='x', rotation=45, labelsize=10)  # Rotate x-axis labels
        axes[i].tick_params(axis='y', labelsize=10)  # Adjust y-axis font size
        axes[i].spines['top'].set_visible(False)  # Remove top border
        axes[i].spines['right'].set_visible(False)  # Remove right border
        axes[i].grid(visible=False)  # Remove grid lines
        axes[i].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x)}' if x == int(x) else ''))

    plt.tight_layout()
    plt.show()

    # Step 4: Revert labels back to numbers
    reverse_value_map = {v: k for k, v in value_map.items()}
    df[columns] = df[columns].replace(reverse_value_map).astype('Int64')  # Ensure the columns revert to integers

    return df
```

```python
import os

def overall_plots(df, columns, value_map, figsize=(20, 15), plot_color="#4682B4", output_folder="exports"):
    """
    Generate overall histograms for multiple columns in a grid layout.
    Map values to descriptive labels for better visualization, ensure consistent x-axis order,
    and export the plots to both PDF and PNG formats.

    Parameters:
    - df (DataFrame): The DataFrame to process.
    - columns (list): List of column names to plot.
    - value_map (dict): Mapping of numerical values to descriptive labels.
    - figsize (tuple): Base figure size for the grid of plots.
    - plot_color (str): Hex color code for the plots.
    - output_folder (str): Folder to save the exported plots.

    Returns:
    - DataFrame: The original DataFrame with columns reverted back to numbers.
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    import math
    import matplotlib.ticker as ticker
    from pandas.api.types import CategoricalDtype

    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Step 1: Map the columns to descriptive labels, ignoring unmapped or NA entries
    df_temp = df[columns].astype('object')
    df_mapped = df_temp.applymap(lambda x: value_map.get(x) if x in value_map else None)

    # Define consistent category order
    category_order = ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree']

    # Convert all columns to categorical with the defined order
    for col in columns:
        df_mapped[col] = df_mapped[col].astype(CategoricalDtype(categories=category_order, ordered=True))

    # Step 2: Determine the number of rows and columns for the grid
    n_cols = 2  # Fixed number of columns
    n_rows = math.ceil(len(columns) / n_cols)  # Number of rows needed

    # Dynamically adjust figure size based on rows
    fig_height = max(figsize[1], n_rows * 6)  # Increased row height for more space
    fig_width = max(figsize[0], n_cols * 10)  # Increased width for more space
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(fig_width, fig_height))

    # Flatten axes for easier indexing
    axes = axes.flatten()

    # Step 3: Generate plots for each column
    for i, col in enumerate(columns):
        # Drop NA values for the column being plotted
        column_data = df_mapped[col].dropna()

        # Plot the column
        sns.histplot(column_data, kde=False, discrete=True, ax=axes[i], color=plot_color)
        axes[i].set_title(f"Overall {col.replace('_', ' ').title()}", fontsize=16)  # Larger title font
        axes[i].set_xlabel('Response', fontsize=14)  # Larger x-axis label font
        axes[i].set_ylabel('Frequency', fontsize=14)  # Larger y-axis label font
        axes[i].tick_params(axis='x', rotation=45, labelsize=12)  # Rotate labels for better fit
        axes[i].tick_params(axis='y', labelsize=12)
        axes[i].spines['top'].set_visible(False)  # Remove top border
        axes[i].spines['right'].set_visible(False)  # Remove right border
        axes[i].grid(False)  # Turn off grid lines for a cleaner look
        axes[i].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x)}' if x == int(x) else ''))

    # Hide any unused subplots in the grid
    for j in range(len(columns), len(axes)):
        axes[j].axis('off')

    # Adjust layout for better readability
    plt.subplots_adjust(hspace=0.5, wspace=0.4)  # Add more space between rows and columns
    plt.tight_layout(pad=5)  # Further increase padding for overall layout

    # Export the figure to PDF and PNG formats
    pdf_path = os.path.join(output_folder, "overall_plots.pdf")
    png_path = os.path.join(output_folder, "overall_plots.png")
    plt.savefig(pdf_path, format="pdf", bbox_inches="tight")
    plt.savefig(png_path, format="png", bbox_inches="tight")
    plt.show()

    print(f"Plots exported to:\nPDF: {pdf_path}\nPNG: {png_path}")

    # Step 4: Revert labels back to numbers
    reverse_value_map = {v: k for k, v in value_map.items()}
    df[columns] = df[columns].replace(reverse_value_map).astype('Int64')  # Ensure the columns revert to integers

    return df
```

```python
def single_plot(df, column, value_map, figsize=(6, 4.5), plot_color="#4682B4"):
    """
    Generate a histogram for a single column.
    Map values to descriptive labels for better visualization and ignore unmapped/NA entries.

    Parameters:
    - df (DataFrame): The DataFrame to process.
    - column (str): Column name to plot.
    - value_map (dict): Mapping of numerical values to descriptive labels.
    - figsize (tuple): Figure size for the plot.
    - plot_color (str): Hex color code for the plot.

    Returns:
    - DataFrame: The original DataFrame with the column reverted back to numbers.
    """

    # Step 1: Map the column to descriptive labels, ignoring unmapped or NA entries
    column_data = df[column].map(value_map).dropna()

    # Step 2: Plot the histogram
    plt.figure(figsize=figsize)
    sns.histplot(column_data, kde=False, discrete=True, color=plot_color)
    
    # Step 3: Customize the plot
    plt.title(f"Distribution of {column.replace('_', ' ').title()}", fontsize=10)
    plt.xlabel('Response', fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.xticks(rotation=45, fontsize=10)
    plt.yticks(fontsize=10)
    plt.gca().spines['top'].set_visible(False)  # Remove top border
    plt.gca().spines['right'].set_visible(False)  # Remove right border
    plt.gca().grid(visible=False)  # Remove grid lines
    plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x)}' if x == int(x) else ''))

    plt.tight_layout()
    plt.show()

    # Step 4: Revert labels back to numbers
    reverse_value_map = {v: k for k, v in value_map.items()}
    df[column] = df[column].replace(reverse_value_map).astype('Int64')  # Ensure the column reverts to integers

    return df
```

```python
def summarize_visualization_insights(df, columns, value_map):
    """
    Summarize insights from the visualizations into a table format.

    Parameters:
    - df (DataFrame): The dataset containing the variables.
    - columns (list): List of column names to summarize.
    - value_map (dict): Mapping of numerical values to descriptive labels.

    Returns:
    - DataFrame: A summary table with insights.
    """
    import pandas as pd

    summary = []

    for col in columns:
        # Temporarily map values to descriptive labels
        col_data = df[col].map(value_map).dropna()  # Map values and drop NA

        if col_data.empty:  # Skip if no valid data
            continue

        # Determine the dominant response
        dominant_response = col_data.value_counts().idxmax()  # Most frequent response
        total_responses = col_data.value_counts(normalize=True) * 100  # Percentage distribution
        skewed_positive = total_responses.get('Agree', 0) + total_responses.get('Strongly Agree', 0)
        skewed_negative = total_responses.get('Disagree', 0) + total_responses.get('Strongly Disagree', 0)

        # Summarize trends and barriers
        notable_trends = (
            "Strong positive skew" if skewed_positive > 60 else
            "Strong negative skew" if skewed_negative > 60 else
            "Balanced distribution"
        )
        potential_barriers = (
            "Neutral responses dominate" if total_responses.get('Neutral', 0) > 30 else
            "Significant disagreement" if skewed_negative > 30 else
            "Minimal barriers"
        )

        # Append to summary
        summary.append({
            "Variable": col.replace("_", " ").title(),
            "Dominant Response": dominant_response,
            "Notable Trends": notable_trends,
            "Potential Barriers": potential_barriers
        })

    # Convert to DataFrame
    summary_df = pd.DataFrame(summary)
    return summary_df
```

```python
# Define value map
value_map_general = {
    1: 'Strongly Disagree',
    2: 'Disagree',
    3: 'Neutral',
    4: 'Agree',
    5: 'Strongly Agree'
}

# List of columns to process
columns_to_process = ['detection_at_birth',
       'scd_longevity', 'early_detection_importance', 'need_immediate_result',
       'enough_poct_info', 'poct_result_speed', 'screening_recommendation',
       'poct_recommendation', 'dbs_for_validation', 'proper_treatment']

# Generate the summary table
summary_table = summarize_visualization_insights(df, columns_to_process, value_map_general)

# # Display the summary
# import ace_tools as tools; tools.display_dataframe_to_user(name="Visualization Insights Summary", 
#                                                            dataframe=summary_table)
```

```python
def create_violin_plots(df, x_col, y_col, value_map=None, figsize=(8, 6)):
    """
    Create violin plots for a selected column grouped by a categorical variable with a modern aesthetic.
    
    Parameters:
    - df (DataFrame): The DataFrame to process.
    - x_col (str): Column name to use for the x-axis (categorical grouping).
    - y_col (str): Column name to use for the y-axis (numeric or ordinal values).
    - value_map (dict): Optional mapping of numeric values to descriptive labels (used for y-axis ticks).
    - figsize (tuple): Size of the plot.
    
    Returns:
    - None: Displays the plot.
    """
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    df['country'] = df['country'].str.capitalize()

    # Create the violin plot
    plt.figure(figsize=figsize)
    sns.set_theme(style="whitegrid")  # Modern white grid theme
    sns.violinplot(data=df, x=x_col, y=y_col, palette="coolwarm", inner="quartile", linewidth=1.2)

    # Update y-axis ticks with descriptive labels if value_map is provided
    if value_map:
        plt.yticks(
            ticks=list(value_map.keys()),  # Numeric values
            labels=list(value_map.values()),  # Corresponding descriptive labels
            fontsize=10
        )

    # Remove top and right borders for a cleaner look
    sns.despine()

    # Add labels and title
    plt.title(f"Distribution of {y_col.replace('_', ' ').title()} by {x_col.replace('_', ' ').title()}", fontsize=11)
    plt.xlabel(x_col.replace('_', ' ').title(), fontsize=10)
    plt.ylabel(y_col.replace('_', ' ').title(), fontsize=10)

    # Adjust x-axis label font size
    plt.xticks(rotation=45, fontsize=10)

    # Tight layout for better spacing
    plt.tight_layout()
    plt.show()
```

```python
def create_violin_plots_by_country_and_gender(df, x_col, y_col, hue_col, value_map=None, figsize=(8,6)):
    """
    Create violin plots for a selected column grouped by a categorical variable (e.g., country),
    and separated by gender (or another grouping variable) using the hue parameter.
    
    Parameters:
    - df (DataFrame): The DataFrame to process.
    - x_col (str): Column name to use for the x-axis (categorical grouping, e.g., country).
    - y_col (str): Column name to use for the y-axis (numeric or ordinal values).
    - hue_col (str): Column name for additional grouping (e.g., gender).
    - value_map (dict): Optional mapping of numeric values to descriptive labels (used for y-axis ticks).
    - figsize (tuple): Size of the plot.
    
    Returns:
    - None: Displays the plot.
    """
    import seaborn as sns
    import matplotlib.pyplot as plt

    # Map the hue column to descriptive labels (e.g., Male/Female)
    df_temp = df.copy()
    df_temp[hue_col] = df_temp[hue_col].replace({1: 'Male', 0: 'Female'})

    # Create the violin plot
    plt.figure(figsize=figsize)
    sns.violinplot(data=df_temp, x=x_col, y=y_col, hue=hue_col, split=True, palette="pastel", inner="quartile")

    # Update y-axis ticks with descriptive labels if value_map is provided
    if value_map:
        plt.yticks(
            ticks=list(value_map.keys()),  # Numeric values
            labels=list(value_map.values()),  # Corresponding descriptive labels
            fontsize=10
        )

    # Add labels and title
    plt.title(f"Distribution of {y_col.replace('_', ' ').title()} by {x_col.replace('_', ' ').title()} and Gender", fontsize=10)
    plt.xlabel(x_col.replace('_', ' ').title(), fontsize=10)
    plt.ylabel(y_col.replace('_', ' ').title(), fontsize=10)

    # Adjust x-axis label font size
    plt.xticks(rotation=45, fontsize=10)

    # Remove top and right borders for a cleaner look
    sns.despine()

    plt.tight_layout()
    plt.show()
```

```python
def plot_diverging_stacked_bars(df, columns, value_map, figsize=(8, 4)):
    """
    Generate diverging stacked bar charts for multiple Likert scale columns, excluding NaN entries.

    Parameters:
    - df (DataFrame): The DataFrame to process.
    - columns (list): List of Likert scale columns to plot.
    - value_map (dict): Mapping of numeric Likert scale values to descriptive text.
    - figsize (tuple): Size of the plot for each column.

    Returns:
    - None: Displays the plots.
    """
    import seaborn as sns
    import matplotlib.pyplot as plt

    for col in columns:
        # Exclude NaN values for the current column
        df_filtered = df[df[col].notna()]

        # Create a pivot table for percentage responses by country
        summary = (
            df_filtered.groupby(['country', col])[col]
            .count()
            .unstack()
            .fillna(0)
        )
        
        # Normalize to percentage
        summary = (summary.T / summary.T.sum()).T * 100

        # Map the Likert scale values to words for the column labels
        summary.columns = summary.columns.map(value_map)

        # Plot diverging bar chart
        fig, ax = plt.subplots(figsize=figsize)
        summary.plot(kind='barh', stacked=True, color=sns.color_palette("coolwarm", n_colors=len(value_map)), ax=ax)

        # Customize the plot aesthetics
        plt.title(f"Diverging Stacked Bar Chart of {col.replace('_', ' ').title()} by Country", fontsize=11)
        plt.xlabel("Percentage (%)", fontsize=10)
        plt.ylabel("Country", fontsize=10)
        plt.legend(
            title="Likert Responses",
            bbox_to_anchor=(1.05, 1),
            loc='upper left',
            fontsize=9,
            title_fontsize=10
        )
        plt.tight_layout()
        sns.despine(left=True, bottom=True)  # Remove unnecessary borders
        plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_violin_plots_by_gender(df, questions, gender_col='sex_guardian', figsize=(7, 3)):
    """
    Plot violin plots for each questionnaire question, comparing male and female responses.

    Parameters:
    - df (DataFrame): The DataFrame containing questionnaire data.
    - questions (dict): Dictionary mapping column names to full questions.
    - gender_col (str): Column indicating gender (e.g., 1 = Male, 0 = Female).
    - figsize (tuple): Size of each violin plot.

    Returns:
    - None: Displays the plots.
    """
    # Map gender column for clearer labels
    df_temp = df.copy()
    df_temp[gender_col] = df_temp[gender_col].replace({1: 'Male', 0: 'Female'})

    for col, question in questions.items():
        plt.figure(figsize=figsize)
        sns.violinplot(data=df_temp, x=gender_col, y=col, palette='pastel', inner='quartile', linewidth=1)
        plt.title(f"{question}\n(Violin Plot by Gender)", fontsize=10)
        plt.xlabel("Gender", fontsize=10)
        plt.ylabel("Response", fontsize=10)
        plt.xticks(fontsize=9)
        plt.yticks([1, 2, 3, 4, 5], ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree'], fontsize=9)
        sns.despine()
        plt.tight_layout()
        plt.show()
```

```python
# Define questions mapping
questions = {
    "detection_at_birth": "Sickle cell disease (SCD) can be detected at birth",
    "scd_longevity": "People with SCD can live long lives",
    "early_detection_importance": "Early detection of SCD is needed to provide adequate care for the affected child",
    "need_immediate_result": "It is important that a test for SCD for my child produces the result immediately",
    "enough_poct_info": "I have access to all the information I need on point of care tests for SCD",
    "poct_result_speed": "The result of a point of care test for SCD is available faster than for other screening methods",
    "screening_recommendation": "I would recommend the screening of babies for SCD at other healthcare facilities",
    "poct_recommendation": "I would recommend the use of point of care tests for SCD at other healthcare facilities",
    "dbs_for_validation": "Alongside the point of care test, it is acceptable for a dried blood spot sample to be collected for validation",
    "proper_treatment": "If someone tests positive for SCD at this healthcare facility it is important that they seek proper medical treatment, either here or at another facility",
    "scd_in_the_family": "Do you have any family member with SCD?",
    "registry_willingness": "If your baby tests positive, would you like to enrol the baby in SCD registry?"
}

# Call the function
plot_violin_plots_by_gender(df, questions)
```

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def plot_all_violin_plots_with_legends(
    df, questions, country_col='country', gender_col='sex_guardian', 
    export_folder='exports', figsize=(14, 6)
):
    """
    Generate violin plots for all questions, grouped by country and gender, save them to a single PNG and PDF file,
    and display them in the code output. Each plot includes its own legend.

    Parameters:
    - df (DataFrame): The DataFrame containing questionnaire data.
    - questions (dict): Dictionary mapping column names to full questions.
    - country_col (str): Column indicating the country.
    - gender_col (str): Column indicating gender (e.g., 1 = Male, 0 = Female).
    - export_folder (str): Folder where the combined plots will be saved.
    - figsize (tuple): Size of each individual plot.

    Returns:
    - None: Saves all plots to a single PNG and PDF file, and prints plots to the output.
    """
    # Ensure the export folder exists
    os.makedirs(export_folder, exist_ok=True)

    # Initialize the PDF file for saving
    pdf_file_path = os.path.join(export_folder, "all_violin_plots_with_legends.pdf")
    png_file_path = os.path.join(export_folder, "all_violin_plots_with_legends.png")

    # Open PDF for multi-page export
    pdf_pages = PdfPages(pdf_file_path)

    # Map gender column for clearer labels
    df_temp = df.copy()
    df_temp[gender_col] = df_temp[gender_col].replace({1: 'Male', 0: 'Female'})

    # Generate a unique palette for each country
    unique_countries = df_temp[country_col].unique()
    country_palette = sns.color_palette("tab10", len(unique_countries))
    country_colors = dict(zip(unique_countries, country_palette))

    # Create a figure for all questions
    num_questions = len(questions)
    fig, axes = plt.subplots(num_questions, 1, figsize=(figsize[0], figsize[1] * num_questions), constrained_layout=True)

    if num_questions == 1:
        axes = [axes]  # Ensure axes is iterable if there's only one question

    for ax, (col, question) in zip(axes, questions.items()):
        # Create an x-tick order with "Male" and "Female" under each country
        df_temp['country_gender'] = df_temp[country_col] + " - " + df_temp[gender_col]
        x_order = [
            f"{country} - Male" if gender == 'Male' else f"{country} - Female"
            for country in unique_countries
            for gender in ['Male', 'Female']
        ]

        # Map colors to violins (same color for both genders in the same country)
        palette = {f"{country} - Male": color for country, color in country_colors.items()}
        palette.update({f"{country} - Female": color for country, color in country_colors.items()})

        # Create the violin plot
        sns.violinplot(
            data=df_temp,
            x='country_gender',
            y=col,
            palette=palette,
            inner='quartile',
            linewidth=1,
            order=x_order,
            ax=ax
        )

        # Customize the plot
        ax.set_title(f"{question}\n(Violin Plot by Country and Gender)", fontsize=12, weight="bold")
        ax.set_xlabel("Gender", fontsize=10)
        ax.set_ylabel("Response", fontsize=10)

        # Customize x-ticks to show "Male" and "Female"
        ax.set_xticks(range(len(x_order)))
        ax.set_xticklabels(
            ['Male' if 'Male' in label else 'Female' for label in x_order],
            rotation=0,
            fontsize=9
        )

        # Add a legend for each plot
        legend_elements = [
            plt.Line2D([0], [0], color=color, lw=4, label=country)
            for country, color in country_colors.items()
        ]
        ax.legend(
            handles=legend_elements,
            title="Country",
            bbox_to_anchor=(1.05, 1),
            loc='upper left',
            fontsize=9,
            title_fontsize=10,
            frameon=False  # Remove the legend border
        )

        # Adjust y-axis labels
        ax.set_yticks([1, 2, 3, 4, 5])
        ax.set_yticklabels(
            ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree'],
            fontsize=9
        )
        sns.despine(ax=ax)

    # Adjust layout to include legends
    plt.tight_layout(rect=[0, 0, 0.85, 1])

    # Save PNG file with legends
    plt.savefig(png_file_path, format='png', dpi=300, bbox_inches='tight')

    # Save PDF file with legends
    pdf_pages.savefig(fig, bbox_inches='tight')
    pdf_pages.close()

    # Display plots inline in the output
    plt.show()

    # Close the figure after saving and displaying
    plt.close(fig)

    print(f"Plots saved successfully:\n- PNG: {png_file_path}\n- PDF: {pdf_file_path}")
```
